```python
import requests
from bs4 import BeautifulSoup
```

```python
def crawl(max):
    page=1
    while(page<=max):
        v="https://internshala.com/internships/computer%20science-internship/page-" + str(page);
        code = requests.get(v)
        plain = code.text
        soup = BeautifulSoup(plain)
        for link in soup.findAll('a',{'class':'link_display_like_text'}):
            title = link.string
            print(title)
        page=page+1;
crawl(2)
```
